# Scraping Dataset Ulasan Aplikasi Dana pada Google Play Store

```python
scrapreview = reviews_all(
    'id.dana', #IDnya aplikasi Dana
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=10000 #10000 ulasan yang diambil
)
```

# Loading Dataset

```python
#menyimpan ulasan dalam file csv
with open('ulasan_aplikasi_dana.csv', mode='w', newline='', encoding='utf-8') as file:
  writer = csv.writer(file)
  writer.writerow(['Review'])
  for review in scrapreview:
    writer.writerow([review['content']])

app_reviews_df = pd.DataFrame(scrapreview)
app_reviews_df.shape
app_reviews_df.head()
app_reviews_df.to_csv('/content/drive/MyDrive/Colab Notebooks/projekdicoding/proyek_analisis_sentimen/ulasan_aplikasi_dana.csv', index=False)

#membuat dataframe dari hasil scrapreview
dana_reviews_df = pd.DataFrame(scrapreview)
```
